```python
import kagglehub
import os

# Define the target directory
download_dir = './dataset'
os.makedirs(download_dir, exist_ok=True)

download_path = kagglehub.dataset_download(
    'matthewjansen/ucf101-action-recognition',
)

print(f"Dataset downloaded to: {download_path}")
```

```python
import pandas as pd
```

```python
from pathlib import Path
from random import shuffle

classes = ['Diving', 'Drumming', 'JugglingBalls']

train_df = pd.DataFrame(
    columns=['clip_name', 'clip_path', 'label', 'encoded_label']
)
val_df = pd.DataFrame(
    columns=['clip_name', 'clip_path', 'label', 'encoded_label']
)
test_df = pd.DataFrame(
    columns=['clip_name', 'clip_path', 'label', 'encoded_label']
)

def get_df(files, label, encoded_label):
    labels = [label] * len(files)
    encoded_labels = [encoded_label] * len(files)
    clip_with_path = [f'./dataset/{label}/{file}' for file in files]
    return pd.DataFrame(
        columns=['clip_name', 'clip_path', 'label', 'encoded_label'],
        data=zip(files, clip_with_path, labels, encoded_labels)
    )

for idx, cls in enumerate(classes):
    directory_path = Path('./dataset/' + cls + '/')
    print("-----------------------------")
    print(directory_path)
    print(f"Total number of images: {len(os.listdir(directory_path))}")
    files = os.listdir(directory_path)
    shuffle(files)
    train_files = files[0:79]
    test_files = files[80:89]
    val_files = files[90:99]
    t1_df = get_df(train_files, cls, idx)
    t2_df = get_df(test_files, cls, idx)
    t3_df = get_df(val_files, cls, idx)
    train_df = pd.concat([train_df, t1_df])
    test_df = pd.concat([test_df, t2_df])
    val_df = pd.concat([val_df, t3_df])

train_df.to_csv('./dataset/splits/train.csv', index=True)
val_df.to_csv('./dataset/splits/validation.csv', index=True)
test_df.to_csv('./dataset/splits/test.csv', index=True)



```
